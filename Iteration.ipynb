```python
for i in range(5):   
    x,y=int(input()),int(input())#x,y--->15,45
    l=list()
    if x<y:
        for i in range(2,x+1):
            if x%i==0 and y%i==0:
                l.append(i)
    else:
        for i in range(2,y+1):
            if x%i==0 and y%i==0:
                l.append(i)


    if len(l)>0:
        print(l[-1])
    else:
        print('No Common Factor')
```

```python
#Guessing Game---->
#computer Guess a number
#User guess computer's number in five chanches
#If user guess greater number ---->think small
#If user guess smaller number---->think big
#if user guess correct----->Won the match---->break
```

```python
import random
import time
com_guess=random.randint(1,50)
chance=1
while chance<=5:
    user_guess=int(input('Enter a number between 1 to 50'))
    if user_guess<1 or user_guess>50:
        print('please enter number between 1 to 50')
        continue
    elif user_guess<com_guess:
        print('Think Big....')
    elif user_guess>com_guess:
        print('Be in Limits and Think Small')
    else:
        print('Congratulation')
        print('You have won the match')
        break
    chance+=1
else:
    print('You are such a looser....')
    for i in 'Better luck next time':
        time.sleep(0.5)
        print(i,end='')
```
